# primeFlex syntax
1. flex
默认： flex-row
位置：
* if flex-row， main axis is X，justify 是子件在x上的位置。align 是在y上的位置。
if flex-coloumn， 相反。
* align-self 指子件自己。 align-self-start 位置：auto，start，center，end，baseline，stetch
* justify-content 内容位置 start, end, space between,around(items around by space),evenly
* align-content 内容位置 start, end, space between,around(items around by space),evenly
* align-item 内容位置 stetch,start, end, center baseline
