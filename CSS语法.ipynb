# css syntax
1. space  one in another    
` li a` : all a inside li

2. comma : and    
`li,a`: li and a

3. : status   
`li:hover` : 当鼠标 悬停 在 li 上时

4. space and  :  
`ul li:hover` : 鼠标 悬停 在 ul 里面的 li 上时

5. 直接子孫 >   
`ul li:hover>ul` : 鼠标悬停在 li 上时，显示它直接子级的 ul（下拉菜单）

6. dot .    
 class `li.selected` :

7. 多個dot   
`li.selected.active` 同一个 li 同时有 selected + active

8. PrimeNG類  selected 这个 class 不是你自己写的，是 PrimeNG / Angular / 路由 自动加上去的！
9. &    
可读写法 nav with class primary-navigation
    ```
    nav {
      background: white;
      
      &.primary-navigation {
        background: blue;
      }
    ```

> 没有& 表示 nav 里 含有 primary-navigation 类的元素。
